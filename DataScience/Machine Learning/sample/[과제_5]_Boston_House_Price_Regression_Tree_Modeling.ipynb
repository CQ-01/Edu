# Boston House Price
* <font color = 'blue'>'boston.csv' (506 * 14)</font>

 - CRIM: 범죄 발생률
 - ZN: 25,000평방피트를 초과하는 거주 지역의 비율
 -  INDUS: 비소매 상업 지역 비율
 - CHAS: 찰스강 인접 여부(강주변 1, 아니면 0)
 - NOX: 일산화질소 농도
 - RM: 거주 목적 방 개수
 - AGE: 1940년 이전 건축된 주택 비율
 - DIS: 보스턴 5대 고용지역까지 거리
 - RAD: 고속도로 접근성
 - TAX: 10,000달러당 재산세율
 - PTRATIO: 교사와 학생 수 비율
 - B: 흑인 거주 비율
 - LSTAT: 하위 계층 비율
 - <font color = 'red'>PRICE: 주택 가격 -> 'y'

```python
import warnings
warnings.filterwarnings('ignore')
```

# I. Data Load
* 'boston.csv' Github에서 읽어오기

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/boston.csv'
DF = pd.read_csv(url)

DF.info()
```

```python
DF.head()
```

# II. 탐색적 데이터 분석(EDA)

> ## 1) 'PRICE' 분포

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(x = 'PRICE',
            data = DF,
            height = 7)
plt.show()
```

* 'PRICE' 평균

```python
DF['PRICE'].mean()
```

> ## 2) 상관계수 ('r')

* 전체 데이터

```python
DF.corr()
```

* HeatMap Visualization

```python
plt.figure(figsize = (12, 12))
sns.heatmap(DF.corr(),
            annot = True,
            cmap = 'Blues',
            fmt = '.2f')
plt.show()
```

> ## 3) 연속형 변수 상관계수 ('r')

* 'PRICE'와 나머지 변수 간 상관계수 내림차순 정렬

```python
DF.corr().loc[:'LSTAT', 'PRICE'].abs().sort_values(ascending = False)
```

* 6위까지 추출

```python
DF_reg = DF.loc[:, ['PRICE', 'LSTAT', 'RM', 'PTRATIO', 'INDUS', 'TAX', 'NOX']].copy()

DF_reg.head()
```

> ## 4) 'PRICE'와 6위까지 변수간 시각화

* 회귀모델 시각화

```python
plt.figure(figsize = (18, 10))

for index, col_names in enumerate(DF_reg.columns[1:]):
  ax1 = plt.subplot(2, 3, index + 1)
  sns.regplot(x = col_names,
              y = DF_reg.columns[0],
              data = DF_reg,
              ax = ax1)

plt.show()
```

* 연속형 변수간 상관관계 ('r') 확인

```python
plt.figure(figsize = (7, 7))
sns.heatmap(DF_reg.loc[:,'LSTAT':].corr(),
            annot = True,
            cmap = 'Reds',
            fmt = '.2f')
plt.show()
```

# III. train_test_split( )

* 8 : 2
 - random_state = 2045

```python
from sklearn.model_selection import train_test_split

X = DF.iloc[:, :-1]
y = DF.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 2045)
print('Train Data : ', X_train.shape, y_train.shape)
print('Test Data : ', X_test.shape, y_test.shape)
```

# IV. Modeling - DecisionTreeRegressor

> ## 1) 모델 생성

```python
from sklearn.tree import DecisionTreeRegressor

DT = DecisionTreeRegressor(random_state = 2045)
DT.fit(X_train, y_train)
```

> ## 2) 모델 평가

### (1) Mean Squared Error (MSE)

```python
import numpy as np
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test, DT.predict(X_test)))
```

### (2) R-Sqaure

```python
from sklearn.metrics import r2_score

r2_score(y_test, DT.predict(X_test))
```

> ## 3) Feature Importance

```python
plt.figure(figsize = (12, 6))
sns.barplot(DT.feature_importances_,
            X_train.columns)
plt.show()
```
